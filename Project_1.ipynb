### Task 1: Basics Practice

```python
# import libraries
import pandas as pd
# read csv file
df = pd.read_csv('employees.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.tail()
```

```python
# change columns name to lowercase 
df.columns = df.columns.str.lower()
df.head()
```

```python
df['salary'].dtype
```

### Task 2: Data Selection & Filtering

```python
#select first 5 rows of age column
df['age'].head(5)
```

```python
# select rows where age > 25
df[df['age'] > 25]
```

```python
# select rows where age> 25  and salary > 60000
df[(df['age'] > 25) & (df['salary'] > 60000)]
```

### Task 3: Data Cleaning

```python
# check missing values 
df.isnull()
```

```python
# fill missing age values by mean age 
df['age'] = df['age'].fillna(df['age'].mean())
df.head()
```

```python
df.drop_duplicates()
```

### Task 4: Aggregation & Grouping

```python
# calculate average salary on basis of department
df.groupby('department')['salary'].mean()
```

```python
df['salary'].describe()
```

### Task 5: Sorting & Indexing

```python
# sort database on the basis of descending order of salary 
df.sort_values(by='salary', ascending=False)
```

```python
df.reset_index(drop= True, inplace= True)
df.head()
```

```python
# set employeeid as index
df.set_index('employeeid', inplace= True)
df.head()
```
