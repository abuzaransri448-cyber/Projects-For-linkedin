```python
#import libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read csv file
df = pd.read_csv('students_performance.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

### 1. Bar Plot

```python
# average score by gender
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)
# bar plot of average score by gender
sns.barplot(x='gender', y='average_score', data=df)
plt.title('Average Score by Gender')
```

#### Insight:
The average score of females is slightly higher than that of males.This indicates a minimal gender-based performance difference in the dataset.

```python
# bar plot of parental level of education vs average score
sns.barplot(x='parental_level_of_education', y='average_score', data=df)
plt.title('Parental Level of Education vs Average Score')
plt.xticks(rotation=45)
```

#### Insights: 
There is a visible positive trend between parental education levels and student performance, showing that higher levels of parental education generally correspond to higher average test scores.

### 2. Histogram

```python
# math score distribution by using histogram
sns.histplot(df['math_score'], bins=10, kde=True)
plt.title('Math Score Distribution')
```

#### Insights:
Most students scored in the higher range (80â€“90), indicating strong overall performance in mathematics.
The distribution is negatively skewed with a visible gap between 65 and 70, showing that low scores are rare in this dataset.

### 3. Box Plot

```python
# boxplot of test_preparation_course vs scores
sns.boxplot(x='test_preparation_course', y='average_score', data=df)
plt.title('Test Preparation Course vs Average Score')
```

#### Insights: 
The "completed" group shows a much wider interquartile range (IQR) and overall spread, indicating that while the course can boost scores, it results in more diverse performance outcomes than the "none" group.

### 4. Correlation heatmap:

```python
# correlation heatmap
corr = df[['math_score', 'reading_score', 'writing_score']].corr()
sns.heatmap(corr)
plt.title('Correlation Heatmap of Scores')
```

### Insights:
Students whose parents have a Master's degree achieve the highest average scores (over 90), while those from High School backgrounds have the lowest (~60).A clear positive trend exists where higher levels of parental education generally correspond to higher student test scores.
