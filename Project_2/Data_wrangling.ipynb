### 1. The Missing Link:

```python
#Find the rows where pickup_borough or dropoff_borough are missing
missing_boroughs = df[df['pickup_borough'].isnull() | df['dropoff_borough'].isnull()]
missing_boroughs.loc[:, ['pickup', 'dropoff','pickup_borough', 'dropoff_borough']].head(10)
```

### 2. Feature Engineering (Duration):

```python
#The dataset has pickup and drop-off times.
# Calculate a new column called trip_duration_mins.
df['trip_duration_mins'] = (df['dropoff'] - df['pickup']).dt.total_seconds() / 60
df.loc[:, ['pickup', 'dropoff', 'trip_duration_mins']].head()
```

### 3. The Logical Filter:

```python
df.columns
```

```python
#Find where the distance is zero but a fare was charged.
#These are data errors that need to be investigated further.
df[(df['distance'] == 0) & (df['fare'] > 0)].sample(10)
```

```python
#Remove them from dataframe
df= df[~((df['distance'] == 0) & (df['fare'] > 0))]
df.sample(5)
```

### 4. Categorical Binning:

```python
#Create a new column called fare_category in df that labels trips as 
#"Low" (< $10), "Medium" ($10-$20), or "High" (> $20).
def categorize_fare(fare):
    if fare < 10:
        return 'Low'
    elif 10 <= fare <= 20:
        return 'Medium'
    else:
        return 'High'
```

```python
# Apply the function to create the new column
df['fare_category'] = df['fare'].apply(categorize_fare)
df.loc[:, ['fare', 'fare_category']].sample(10)
```
